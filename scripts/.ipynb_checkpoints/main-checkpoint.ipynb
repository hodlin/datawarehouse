## Data section

```python
import pandas as pd
import datetime as dt
import numpy as np
```

```python
data = pd.read_csv('../data/transformed_dec_duration.csv', sep=',', #index_col=0,
                   dtype={'Jobname': np.str, 'Startdatetime': dt.datetime, 'Enddatetime': dt.datetime, 
                           'Starttime': np.float, 'Endtime': np.float, 'Jobduration (s)': np.int})
```

```python
format = "%Y-%m-%d %H:%M:%S"
data['Startdatetime'] = pd.to_datetime(data['Startdatetime'], format=format)
data['Enddatetime'] = pd.to_datetime(data['Enddatetime'], format=format)
```

```python

# print(data)
```

```python
unique_jobs_10 = data['Jobname'].value_counts()[:10].index.values
unique_jobs_10
```

```python
top_10 = [data[data.Jobname == unique_jobs_10[i]].sort_values('Startdatetime') for i in range(10)]
```

```python
start_1 = top_10[0].Startdatetime.min()
end_1 = top_10[0].Startdatetime.max()
```

```python
end_1 - start_1 
```

```python
for job in top_10:
    job['Startdate'] = job['Startdatetime'].apply(lambda dt: dt.date().day)
# unique_date = task_1['Startdate'].value_counts()
# unique_date
```

```python
top_10_agg = [job.groupby(by='Startdate').mean() for job in top_10]

# task_2_agg = task_2.groupby(by='Startdate').mean()
# task_3_agg = task_3.groupby(by='Startdate').mean()
# task_4_agg = task_4.groupby(by='Startdate').mean()
# task_5_agg = task_5.groupby(by='Startdate').mean()
```

```python
top_10_agg[0]
```

```python
data['Startdate'] = data['Startdatetime'].apply(lambda dt: dt.date().day)
print(data['Startdate'][:10])
```

```python
data['Startweekday'] = data['Startdatetime'].apply(lambda dt: dt.date().isoweekday())
print(data['Startweekday'][:10])
```

```python
# print(data[:10])
data_weekly = data.groupby(by='Startweekday').count()
print(data_weekly[:10])
```

## Plotting section

```python
import matplotlib
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.figure(1)
plt.figure(figsize=(15, 7.5), dpi= 160, facecolor='w', edgecolor='k')

plt.subplot(2, 1, 1)
plt.hist(data['Starttime'], bins=24, range=(0, 24), color='green')
plt.title('Starttime')
plt.xlabel('Time, h')
plt.ylabel('Number of observation')
x1, x2, y1, y2 = plt.axis()
plt.axis((0, 24, y1, y2))

plt.subplot(2, 1, 2)
plt.hist(data['Endtime'], bins=24, range=(0, 24), color='red')
plt.title('Endtime')
plt.xlabel('Time, h')
plt.ylabel('Number of observations')
x1, x2, y1, y2 = plt.axis()
plt.axis((0, 24, y1, y2))

plt.tight_layout()

plt.show()
```

```python
plt.figure(2)
plt.figure(figsize=(15, 7.5), dpi= 160, facecolor='w', edgecolor='k')
    
n, bins, patches = plt.hist(data['Startdate'], bins=31, color='green')
plt.title('Jobs daily distribution')
plt.xlabel('Day')
plt.ylabel('Number of jobs')
x1, x2, y1, y2 = plt.axis()
plt.axis((0, 32, y1, y2))

plt.tight_layout()
    
plt.show()
# print(sum(n))
```

```python
plt.figure(3)
plt.figure(figsize=(10, 5), dpi= 160, facecolor='w', edgecolor='k')
    
plt.bar(left=data_weekly.index, height=data_weekly['Jobname'], color='green', align='center')
plt.title('Jobs weekly distribution')
plt.xlabel('Day')
plt.ylabel('Number of jobs')
x1, x2, y1, y2 = plt.axis()
plt.axis((0, 8, y1, y2))
plt.xticks(range(1, 9), ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])

# plt.tight_layout()

plt.show()
```
